# Data Visualization - Distribution Analysis:

```python
# Visualize the distribution of 'Sentiment' and 'Confidence Score' using histograms and box plots:

import matplotlib.pyplot as plt

plt.hist(data['Sentiment'], bins=10, edgecolor='black')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.title('Sentiment Distribution')
plt.show()

plt.boxplot(data['Confidence Score'])
plt.ylabel('Confidence Score')
plt.title('Confidence Score Distribution')
plt.show()
```

```python
# Analyze the distribution of 'Source' using a bar plot:
#Bar Plot
plt.bar(source_counts.index, source_counts.values)
plt.xlabel('Source')
plt.ylabel('Count')
plt.title('Source Distribution')
plt.xticks(rotation=90)
plt.show()
```

# Temporal Analysis:

```python
data.head()
```

```python
data_types = data.dtypes
```

```python
data_types
```

```python
data['DateTime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'])
```

```python
data
```

```python
data.drop(columns=['Date', 'Time'], inplace=True)
```

```python
data
```

```python
# a. Calculate the average "Conference Score" per day:

daily_average = data.groupby(data['DateTime'].dt.date)['Confidence Score'].mean()
print(daily_average)
```

```python
# b. Calculate the maximum "Conference Score" per month:

monthly_max = data.groupby(data['DateTime'].dt.to_period('M'))['Confidence Score'].max()
print(monthly_max)
```

```python
# c. Calculate the number of entries per hour:

entries_per_hour = data.groupby(data['DateTime'].dt.hour).size()
print(entries_per_hour)
```

```python
# d. Plot the trend of "Confidence Score" over time:

data.plot(x='DateTime', y='Confidence Score', marker='o', linestyle='-', color='c')

plt.xlabel('DateTime')
plt.ylabel('Confidence Score')
plt.title('Confidence Score over Time')
plt.xticks(rotation=45, ha='center')
plt.grid(which='major', linestyle='--', color='gray', linewidth=0.5, alpha=0.7)
plt.show()


```

Sentiment Analysis

```python
# .str.strip() method is applied to the 'sentiment' column, which removes any leading or trailing whitespaces. Thats why sometimes show NaN values
data['Sentiment'] = data['Sentiment'].str.strip()

# Define a mapping for sentiment values
sentiment_mapping = {'Positive': 1, 'Negative': -1}

# Apply the mapping to the 'Sentiment' column
data['Sentiment_Mapped'] = data['Sentiment'].map(sentiment_mapping)
data


```

```python
sentiment_counts = data['Sentiment'].value_counts()
plt.bar(sentiment_counts.index, sentiment_counts.values)
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Distribution')
plt.show()
```

# Text Analysis

```python

# Before performing text analysis, you may want to preprocess the 'Text' column by removing stop words, converting to lowercase, and tokenizing the text.
# Create a word cloud to visualize the most frequent words:

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import sys, os
os.chdir(sys.path[0])

text = open('modified_feelings.csv', mode='r', encoding='utf-8' ).read()
stopwords = STOPWORDS

# print(stopwords)
print(len(stopwords))

wc = WordCloud(
background_color='white', stopwords=stopwords,
height = 600,
width=400
)
wc.generate(text)

wc.to_file('wordcloud_output2.png')
```
